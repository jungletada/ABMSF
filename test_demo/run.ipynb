```python
all_wealth = []
# This runs the model 100 times, each model executing 30 steps.
for _ in range(100):
    # Run the model
    model = MoneyModel(10)
    for _ in range(30):
        model.step()

    # Store the results
    for agent in model.agents:
        all_wealth.append(agent.wealth)

# Use seaborn
g = sns.histplot(all_wealth, discrete=True)
g.set(title="Wealth distribution", xlabel="Wealth", ylabel="number of agents");
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 已知数据（单位：人民币元）
mean_income = 36883  # 平均可支配收入
median_income = 31370  # 中位数可支配收入

# 计算对数正态分布参数
mu = np.log(median_income)  # 根据中位数计算 μ
sigma_squared = 2 * (np.log(mean_income) - mu)  # 根据均值和 μ 计算 σ^2
sigma = np.sqrt(sigma_squared)

print(f"Estimated μ (log-mean): {mu:.2f}")
print(f"Estimated σ (log-std): {sigma:.2f}")

# 使用对数正态分布生成模拟收入数据
population_size = 10000  # 模拟人群规模
simulated_income = np.random.lognormal(mean=mu, sigma=sigma, size=population_size)

# 绘制收入分布直方图
plt.figure(figsize=(10, 6))
plt.hist(simulated_income, bins=50, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(mean_income, color='r', linestyle='--', label=f"Mean Income: ¥{mean_income}")
plt.axvline(median_income, color='g', linestyle='-', label=f"Median Income: ¥{median_income}")
plt.xlabel("Annual Disposable Income (RMB)")
plt.ylabel("Number of Individuals")
plt.title("Simulated Income Distribution in China (Log-Normal)")
plt.legend()
plt.grid(True)
plt.show()
```
